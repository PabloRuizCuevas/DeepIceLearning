```python
import matplotlib.pyplot as plt
from scipy import optimize
import numpy as np
```

```python
key="83"
directory="/home/pablo/github/DeepIceLearning/my_code/data_"+key+".npy"
```

```python
charge=np.load(directory)[0]
time=np.load(directory)[1]
width=np.load(directory)[2]
flag=np.load(directory)[3]
charge_density=charge/width
```

```python
def plot(x,y):
    plt.scatter(x,y)
    plt.xlim(9800, 10600)
    plt.title('Double bang event')
    plt.xlabel("time ns")
    plt.ylabel("charge")
    plt.savefig("Double_bang4.png")
```

```python
charge7=charge[flag==7]
time7=time[flag==7]
width7=width[flag==7]
density7=charge7/width7

charge5=charge[flag==5]
time5=time[flag==5]
width5=width[flag==5]
density5=charge5/width5

charge3=charge[flag==3]
time3=time[flag==3]
width3=width[flag==3]
density3=charge3/width3

plot(time7,density7)
plot(time5,density5)
plot(time3,density3)
plt.savefig("Double_bang4.png")
```
