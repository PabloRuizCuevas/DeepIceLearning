```python
threshold=np.linspace(0,1,100)
FPR=[]
BAccuracy=[]
Sensitivity=[]
precision=[]
DOR=[]
for i in threshold:
    pred[predi>i]=1
    pred[predi<=i]=5

    TP=sum((pred==5)&(true==5))
    TN=sum((pred==1)&(true==1))
    FP=sum((pred==5)&(true==1))
    FN=sum((pred==1)&(true==5))
    FPR.append(FP/(FP+TN))
    BAccuracy.append(((TP/(TP+FN))+TN/(TN+FP))/2)
    Sensitivity.append(TP/(TP+FN))
    precision.append(TP/(TP+FP))
    DOR.append(TP*TN/(FN+FP))
```

```python
plt.plot(1-threshold,FPR,label="FPR")                ##How many positives are really positives  #porcentaje de los negativos que son vistos como positivos
plt.plot(1-threshold,BAccuracy,label="BAccuracy")
plt.plot(1-threshold,Sensitivity,label="Sensitivity") ##Percentage of db events identified   (how many positives we detect)
plt.plot(1-threshold,precision,label="precision")     ##percentage of positives that are really positives
plt.plot(1-threshold,1-np.array(precision),label="False Discovery rate")     ##percentaje de los positivos que son falsos
plt.legend()
plt.show()
```

```python
plt.plot(1-threshold,DOR,label="Diagnosis Odds ratio")     ##percentaje de los positivos que son falsos
plt.legend()
plt.show()
```

```python
precision2d=[]
m2=np.linspace(0,600,35)
for m in m2:
    threshold=np.linspace(0.01,1,35)
    precision=[]
    for i in threshold:
        pred[predi>i]=1
        pred[predi<=i]=5
        TP=sum((pred==5)&(true==5)&(length>=m))
        FP=sum((pred==5)&(true==1))
        precision.append(TP/(TP+FP))
    
    precision2d.append(precision)
```
