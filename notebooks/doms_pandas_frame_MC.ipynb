```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from scipy.stats import moment

path="/home/pablo/github/DeepIceLearning/notebooks/data/double_bang_real"
df = pd.read_pickle(path)
```

```python
def nmoment(x, counts, c, n):
    return np.sum(counts*(x-c)**n) / np.sum(counts)
```

```python
df
```

```python
#Calculate all the cuantities you will use:
```

```python
df["tot_charge"]  =df["charge"].map(lambda x: sum(x))
df["first_record"]=df["time"].map(lambda x: min(x))
df["records"]=df["time"].map(lambda x: np.size(x))
df['time_max_charge']  = df[['time','charge']].apply(lambda row: row['time'][np.argmax(row['charge'])] , axis=1)
df['cutted_charge']    = df[['time','charge']].apply(lambda row: row['charge'][row['time'] < min(row['time'])+400] , axis=1)
df['cutted_time']      = df[['time','charge']].apply(lambda row: row['time'][row['time'] < min(row['time'])+400] , axis=1) 
df["normalized_charge"]= df[["cutted_charge"]].apply(lambda row: row["cutted_charge"]/sum(row["cutted_charge"]) , axis=1)
df["normalized_time"]  = df[["cutted_time","first_record"]].apply(lambda row: (row["cutted_time"]-row["first_record"])/(row["cutted_time"][-1]-row["first_record"])-0.5 , axis=1)
df["Skewed"]=    df[["normalized_charge","normalized_time"]].apply(lambda row: nmoment(row["normalized_time"], row["normalized_charge"], 0 , 3), axis=1)
df["Variance"]=    df[["normalized_charge","normalized_time"]].apply(lambda row: nmoment(row["normalized_time"], row["normalized_charge"], 0 , 4), axis=1)
df["Kurtosis"]=    df[["normalized_charge","normalized_time"]].apply(lambda row: nmoment(row["normalized_time"], row["normalized_charge"], 0 , 4), axis=1)
```

```python
def kl_divergence(p, q):
    return np.sum(np.where(p != 0, p * np.log(p / q), 0))
```
