```python
Accuracym=[]
Sensitivity=[]
BalAcc=[]
true2=true.copy()
for i in range(0,100):
    true2[length<i]=1  ##make no_resoluble_db cascades
    
    TP=sum((pred==5)&(true2==5))
    TN=sum((pred==1)&(true2==1))
    FP=sum((pred==5)&(true2==1))
    FN=sum((pred==1)&(true2==5))
    Accuracy = ((TP + TN)/(TP + TN + FP + FN))
    BalAcc.append(((TP/(TP+FN))+TN/(TN+FP))/2)
    Sensitivity.append((TP)/(TP+FN))
    Accuracym.append(Accuracy)
```

```python
plt.plot(range(0,100),Accuracym,label="Acc")
plt.plot(range(0,100),BalAcc,label="BA Acc")
plt.plot(range(0,100),Sensitivity,label="db Sensitivity")
plt.ylabel('P')
plt.xlabel('length more than x meters')
plt.legend()
plt.show()
```

```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

DB=TP+FN
cas=TN+FP

array = [np.array([TP,FN])/DB,np.array([FP,TN])/cas]
df_cm = pd.DataFrame(array, index = [i for i in ["True: DB","Cascade"]],
                  columns = [i for i in  ["Pred: DB","Cascade"]])
plt.figure(figsize = (10,7))

ax=sn.heatmap(df_cm,center=0, annot=True, cmap=sn.diverging_palette(220, 100, as_cmap=True))
ax.set_ylim(0, 2)
```

```python
threshold=np.linspace(0,1,100)
FPR=[]
BAccuracy=[]
Sensitivity=[]
precision=[]
for i in threshold:
    pred[predi>i]=1
    pred[predi<=i]=5

    TP=sum((pred==5)&(true==5))
    TN=sum((pred==1)&(true==1))
    FP=sum((pred==5)&(true==1))
    FN=sum((pred==1)&(true==5))
    FPR.append(FP/(FP+TN))
    BAccuracy.append(((TP/(TP+FN))+TN/(TN+FP))/2)
    Sensitivity.append(TP/(TP+FN))
    precision.append(TP/(TP+FP))
```

```python
plt.plot(1-threshold,FPR)
plt.plot(1-threshold,BAccuracy)
plt.plot(1-threshold,Sensitivity)
plt.plot(1-threshold,precision)
plt.show()
```

```python
precision2d=[]
m2=np.linspace(0,400,20)
for m in m2:
    threshold=np.linspace(0.01,1,30)
    precision=[]
    for i in threshold:
        pred[predi>i]=1
        pred[predi<=i]=5
        TP=sum((pred==5)&(true==5)&(length>=m))
        FP=sum((pred==5)&(true==1))
        precision.append(TP/(TP+FP))
    precision2d.append(np.array(precision))
```

```python
sn.heatmap(precision2d)
plt.show()
```
