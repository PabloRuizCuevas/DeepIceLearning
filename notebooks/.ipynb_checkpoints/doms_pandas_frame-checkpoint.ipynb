```python
def normalize(x,y):
    y=y[x< (min(x)+200)]
    x=x[x< (min(x)+200)]
    y=y/sum(y)
    return x,y
```

```python
x2,y2=normalize(x,y)
x2
```

```python
vals, bins = np.histogram(x2,weights=y2,bins=np.arange(np.min(x2)-10, np.min(x2)+400,7))
mask = vals>0
plt.errorbar((bins[:-1]+bins[1:])[mask]/2, vals[mask], xerr=np.diff(bins)[mask]/2, linestyle="-")
plt.show()
```
