```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from scipy.stats import moment

path="/home/pablo/github/DeepIceLearning/notebooks/data/double_bang_real"
df = pd.read_pickle(path)
```

```python
def nmoment(x, counts, c, n):
    return np.sum(counts*(x-c)**n) / np.sum(counts)
```

```python
df
```

```python
#Calculate all the cuantities you will use:
```

```python
df["tot_charge"]  =df["charge"].map(lambda x: sum(x))
df["first_record"]=df["time"].map(lambda x: min(x))
df['time_max_charge']  = df[['time','charge']].apply(lambda row: row['time'][np.argmax(row['charge'])] , axis=1)
df['cutted_charge']    = df[['time','charge']].apply(lambda row: row['charge'][row['time'] < min(row['time'])+200] , axis=1)
df['cutted_time']      = df[['time','charge']].apply(lambda row: row['time'][row['time'] < min(row['time'])+200] , axis=1) 
df["normalized_charge"]= df[["cutted_charge"]].apply(lambda row: row["cutted_charge"]/sum(row["cutted_charge"]) , axis=1)
df["normalized_time"]  = df[["cutted_time","first_record"]].apply(lambda row: (row["cutted_time"]-row["first_record"])/(row["cutted_time"][-1]-row["first_record"])-0.5 , axis=1)
df["Skewed"]=    df[["normalized_charge","normalized_time"]].apply(lambda row: nmoment(row["normalized_time"], row["normalized_charge"], 0 , 3), axis=1)
```

```python
db_10_20=df[(df["class"]==5)& ((df["lenght"]>=10)&(df["lenght"]<=30) ) & (df["recorded_n"]<=10) ]
cascades=df[(df["class"]==1) & (df["recorded_n"]<=10) ]
```

```python
xd=np.array(db_10_20["time"])
yd=np.array(db_10_20["charge"])
xc=np.array(cascades["time"])
yc=np.array(cascades["charge"])
```

```python
charge_d= np.array([sum(yd[i]) for i in range(xd.size)])
charge_c= np.array([sum(yc[i]) for i in range(xc.size)])
```

```python
plt.hist(charge_d,alpha=0.5)
plt.hist(charge_c,alpha=0.5)
```

```python
def normalize(x,y):
    y=y[x< (min(x)+200)]
    x=x[x< (min(x)+200)]-min(x)
    #y=y/sum(y)
    return x,y
```

```python
charge_threshold=4000
xd=xd[charge_d>charge_threshold]
yd=yd[charge_d>charge_threshold]
charge_d=charge_d[charge_d>charge_threshold]
xc=xc[charge_c>charge_threshold]
yc=yc[charge_c>charge_threshold]
charge_c=charge_c[charge_c>charge_threshold]
```

```python
for i in range(xd.size):
    xd[i],yd[i] = normalize( xd[i],yd[i] )
for i in range(xc.size):  
    xc[i],yc[i] = normalize( xc[i],yc[i] )
```

```python
conditionc=np.array([sum(np.array_split(yc[i],2)[0])>2*sum(np.array_split(yc[i],2)[1]) for i in range(yc.size)])
conditiond=np.array([sum(np.array_split(yd[i],2)[0])>2*sum(np.array_split(yd[i],2)[1]) for i in range(yd.size)])
```

```python
plt.hist(charge_d,alpha=0.5)
plt.hist(charge_c,alpha=0.5)
plt.show()
```

```python
xc=xc[conditionc]
yc=yc[conditionc]
xd=xd[conditiond]
yd=yd[conditiond]
charge_c=[sum(yc[i]) for i in range(yc.size)]
charge_d=[sum(yd[i]) for i in range(yd.size)]
yc=yc/charge_c
yd=yd/charge_d
```
